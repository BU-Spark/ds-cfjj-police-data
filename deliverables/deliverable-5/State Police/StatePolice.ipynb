```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Master.csv")
print(df.Ethnicity.unique())
df.loc[df["Ethnicity"] == "AFRICAN\nAMERICAN", "Ethnicity"] = "AFRICAN AMERICAN"

df.head()
```

```python
total_race = df.groupby(['Race']).size()
fig = plt.figure()

fig.set_figwidth(20)
plt.bar(total_race.index, total_race.values)
plt.xlabel=('Races')
plt.ylabel=('Total FIOs Reported')
plt.title('FIOs by Race')
plt.show()
```

```python
total_ethnic = df.groupby(['Ethnicity']).size()
fig = plt.figure()
fig.set_figwidth(40)
plt.bar(total_ethnic.index, total_ethnic.values)
plt.xlabel=('Races')
plt.ylabel=('Total FIOs Reported')
plt.title('FIOs by Ethnicity')
plt.show()
```

```python
total_age = df.groupby(['Age']).size()
fig = plt.figure()

plt.bar(total_age.index, total_age.values)
plt.xlabel=('Races')
plt.ylabel=('Total FIOs Reported')
plt.title('FIOs by age')
plt.show()
```
