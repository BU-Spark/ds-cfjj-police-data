New Bedford

From NBPD+Validated+Gang+Member+List.csv

```python
df = pd.read_csv('/content/drive/Shareddrives/CS506 Project/deliverables/Deliverable 2/New Bedford/NBPD+Validated+Gang+Member+List.csv')
display(df)
df['Sex'].describe()
```

```python


def to_numerical(df):
    replacements = {'H': 'O', 'I': 'O', 'M': 'O'}
    return df["Race"].replace(replacements)
df['Race'] = to_numerical(df)
total_race = df.groupby(['Race']).size().nlargest()
fig = plt.figure()
fig.patch.set_facecolor('white')
fig.patch.set_alpha(1)
print(total_race.values)
plt.bar(total_race.index, total_race.values, color='blue')
plt.xlabel=('Races')
plt.ylabel=('Total FIOs Reported')
plt.title('Number of FIOs by Race')
plt.show()
```

```python
color = ['blue', 'red', 'green', 'orange']
fig = plt.figure(figsize=(13, 13))
fig.patch.set_facecolor('white')
fig.patch.set_alpha(1)
plot = total_race.plot.pie(subplots=True, autopct='%1.1f%%', colors=color, textprops={'fontsize': 20})
#textprops={'fontsize': 20} inside ax.plot 
plt.legend()
plt.title('Percentage of FIOs based on Race')
```

```python
total_Ethnicity = df.groupby(['Ethnicity']).size()
fig = plt.figure()
fig.patch.set_facecolor('white')
fig.patch.set_alpha(0.3)
plt.bar(total_Ethnicity.index, total_Ethnicity.values, color='blue')
plt.xlabel=('Ethnicity')
plt.ylabel=('Total FIOs Reported')
plt.title('Number of FIOs by Ethnicity')
plt.show()
```

```python
color = ['yellow', 'blue', 'green']
fig = plt.figure(figsize=(13, 13))
fig.patch.set_facecolor('white')
fig.patch.set_alpha(1)
plot = total_Ethnicity.plot.pie(subplots=True, autopct='%1.1f%%', colors=color,textprops={'fontsize': 20})
plt.legend()
plt.title('Percentage of gang based on Ethnicity')
```

```python
total_Age = df.groupby(['Age']).size()
fig = plt.figure()
fig.patch.set_facecolor('white')
fig.patch.set_alpha(0.3)
plt.bar(total_Age.index, total_Age.values, color='blue')
plt.xlabel=('Age')
plt.ylabel=('Total FIOs Reported')
plt.title('Number of FIOs by Age')
plt.show()
```

```python
total_Sex = df.groupby(['Sex']).size()
fig = plt.figure()
fig.patch.set_facecolor('white')
fig.patch.set_alpha(0.3)
plt.bar(total_Sex.index, total_Sex.values, color=["pink",'blue',"green"])
plt.xlabel=('Sex')
plt.ylabel=('Total FIOs Reported')
plt.title('Number of FIOs by Sex')
plt.show()
```
